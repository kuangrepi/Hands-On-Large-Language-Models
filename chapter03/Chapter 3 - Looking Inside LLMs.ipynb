# Loading the LLM

```python
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

# Load model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")

model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)

# Create a pipeline
generator = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    return_full_text=False,
    max_new_tokens=50,
    do_sample=False,
)
```

# The Inputs and Outputs of a Trained Transformer LLM


```python
prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happened."

output = generator(prompt)

print(output[0]['generated_text'])
```

# Choosing a single token from the probability distribution (sampling / decoding)

```python
prompt = "The capital of France is"

# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids

# Tokenize the input prompt
input_ids = input_ids.to("cuda")

# Get the output of the model before the lm_head
model_output = model.model(input_ids)

# Get the output of the lm_head
lm_head_output = model.lm_head(model_output[0])
```

```python
token_id = lm_head_output[0,-1].argmax(-1)
# ä¸€èˆ¬è€Œè¨€ï¼Œè¾“å…¥æ•°æ®ä¼šä»¥æ‰¹é‡çš„å½¢å¼ç»™åˆ°æ¨¡åž‹ï¼Œlm_head_output çš„ç¬¬ä¸€ä¸ªç»´åº¦ä»£è¡¨æ‰¹é‡å¤§å°ï¼ˆbatch_sizeï¼‰ã€‚[0] æ„å‘³ç€ä»Žæ‰¹é‡é‡Œå–å‡ºç¬¬ä¸€ä¸ªæ ·æœ¬ã€‚
# ç¬¬äºŒä¸ªç»´åº¦é€šå¸¸è¡¨ç¤ºåºåˆ—é•¿åº¦ï¼ˆsequence_lengthï¼‰ï¼Œ[-1] è¡¨ç¤ºå–åºåˆ—ä¸­çš„æœ€åŽä¸€ä¸ªæ—¶é—´æ­¥çš„è¾“å‡ºã€‚
# argmax å‡½æ•°çš„ä½œç”¨æ˜¯è¿”å›žå¼ é‡é‡Œæœ€å¤§å€¼æ‰€åœ¨çš„ç´¢å¼•ã€‚-1 è¡¨ç¤ºåœ¨æœ€åŽä¸€ä¸ªç»´åº¦ä¸ŠæŸ¥æ‰¾æœ€å¤§å€¼çš„ç´¢å¼•ã€‚
tokenizer.decode(token_id)
```
