# Downloading and Running An LLM

The first step is to load our model onto the GPU for faster inference. Note that we load the model and tokenizer separately and keep them as such so that we can explore them separately.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happened.<|assistant|>"

# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
# (prompt, return_tensors="pt")ÔºöË∞ÉÁî®ÂàÜËØçÂô®ÂØπ prompt ËøõË°åÂàÜËØçÂ§ÑÁêÜÔºåreturn_tensors="pt" Ë°®Á§∫ËøîÂõû PyTorch Âº†Èáè„ÄÇ
# .input_idsÔºö‰ªéÂàÜËØçÂô®ÁöÑËæìÂá∫‰∏≠ÊèêÂèñ input_idsÔºåÂç≥ÁªèËøáÂàÜËØçÂêéÂæóÂà∞ÁöÑÊï∞Â≠óÂ∫èÂàó„ÄÇ

# Generate the text
generation_output = model.generate(
  input_ids=input_ids, # ÊåáÂÆöÊ®°ÂûãÁöÑËæìÂÖ•ÔºåÂç≥‰πãÂâçÁªèËøáÂàÜËØçÂ§ÑÁêÜÂæóÂà∞ÁöÑÊï∞Â≠óÂ∫èÂàó„ÄÇ
  max_new_tokens=20 # ËÆæÁΩÆÊ®°ÂûãÊúÄÂ§öÁîüÊàê x tokensÔºå‰ΩÜÊòØ‰ºº‰πé‰∏çËÉΩ‰øùËØÅËøôÊÆµËØùËÉΩÁªìÂ∞æ
)

# Print the output
print(tokenizer.decode(generation_output[0])) # tokenizer.decodeÔºöÂ∞ÜÁîüÊàêÁöÑÊï∞Â≠óÂ∫èÂàóÔºàinput_idsÔºâËΩ¨Êç¢ÂõûÊñáÊú¨Â≠óÁ¨¶‰∏≤„ÄÇ
```

# Comparing Trained LLM Tokenizers


```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# colors_list ÊòØ‰∏Ä‰∏™Â≠óÁ¨¶‰∏≤ÂàóË°®ÔºåÊØè‰∏™Â≠óÁ¨¶‰∏≤‰ª£Ë°®‰∏ÄÁßç RGB È¢úËâ≤ÂÄº„ÄÇËøô‰∫õÈ¢úËâ≤ÂÄºÂ∞ÜÁî®‰∫éÂú®ÁªàÁ´Ø‰∏≠‰∏∫‰∏çÂêåÁöÑËØçÂÖÉÊ∑ªÂä†ËÉåÊôØÈ¢úËâ≤Ôºå‰ª•ËææÂà∞ÂèØËßÜÂåñÁöÑÊïàÊûú„ÄÇ
colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(token_ids):
        print(
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
        )
  #  f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m'ÊòØ‰∏Ä‰∏™ ANSI ËΩ¨‰πâÂ∫èÂàóÔºåÁî®‰∫éËÆæÁΩÆÁªàÁ´ØÊñáÊú¨ÁöÑÊ†∑Âºè„ÄÇ
  #  ÂÖ∂‰∏≠ 0 Ë°®Á§∫ÈáçÁΩÆÊâÄÊúâÂ±ûÊÄßÔºå30 Ë°®Á§∫ÂâçÊôØËâ≤‰∏∫ÈªëËâ≤Ôºå48;2 Ë°®Á§∫‰ΩøÁî® RGB È¢úËâ≤ËÆæÁΩÆËÉåÊôØËâ≤Ôºå
  #  {colors_list[idx % len(colors_list)]} ÊòØ‰ªé colors_list ‰∏≠Âæ™ÁéØÈÄâÂèñÁöÑ RGB È¢úËâ≤ÂÄº„ÄÇ
  #  '\x1b[0m'ÔºöÂè¶‰∏Ä‰∏™ ANSI ËΩ¨‰πâÂ∫èÂàóÔºåÁî®‰∫éÈáçÁΩÆÁªàÁ´ØÊñáÊú¨ÁöÑÊ†∑ÂºèÔºåÁ°Æ‰øùÂêéÁª≠ËæìÂá∫‰∏çÂèó‰πãÂâçÊ†∑ÂºèÁöÑÂΩ±Âìç„ÄÇ
  #  end=' 'ÔºöËÆæÁΩÆ print ÂáΩÊï∞ÁöÑÁªìÊùüÁ¨¶‰∏∫Á©∫Ê†ºÔºå‰ΩøÂæóÊØè‰∏™ËØçÂÖÉÂú®Âêå‰∏ÄË°åËæìÂá∫ÔºåÁî®Á©∫Ê†ºÂàÜÈöî„ÄÇ
```

```python
text = """
English and CAPITALIZATION
üéµ È∏ü
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*50=600
""" # ËøôÈáåÊúâÂêÑÁßçtokens
```

```python
show_tokens(text, "bert-base-uncased") # ‰∏çÂå∫ÂàÜÂ§ßÂ∞èÂÜô
```

```python
show_tokens(text, "bert-base-cased") # Âå∫ÂàÜÂ§ßÂ∞èÂÜô
```

```python
show_tokens(text, "gpt2")
```

```python
show_tokens(text, "google/flan-t5-small")
```

```python
# You need to request access before being able to use this tokenizer
show_tokens(text, "bigcode/starcoder2-15b")
```

# Contextualized Word Embeddings From a Language Model (Like BERT)
# Êù•Ëá™ËØ≠Ë®ÄÊ®°ÂûãÁöÑ‰∏ä‰∏ãÊñáÂåñËØçÂµåÂÖ•ÔºàÂ¶Ç BERTÔºâ

```python
from transformers import AutoModel, AutoTokenizer

# Load a tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-base")

# Load a language model
model = AutoModel.from_pretrained("microsoft/deberta-v3-xsmall")

# Tokenize the sentence
tokens = tokenizer('Hello world', return_tensors='pt') # return_tensors='pt' Ë°®Á§∫ËøîÂõû PyTorch Âº†ÈáèÊ†ºÂºèÁöÑÂàÜËØçÁªìÊûú„ÄÇ

# Process the tokens
output = model(**tokens)[0]
# **tokens ÊòØ Python ‰∏≠ÁöÑËß£ÂåÖÊìç‰ΩúÔºåÂ∞Ü tokens Â≠óÂÖ∏‰∏≠ÁöÑÈîÆÂÄºÂØπËß£ÂåÖ‰∏∫ÂÖ≥ÈîÆÂ≠óÂèÇÊï∞‰º†ÈÄíÁªô model„ÄÇËøôÊ†∑Ôºåmodel Â∞±ÂèØ‰ª•Êé•Êî∂ input_ids„ÄÅattention_mask Á≠âÂøÖË¶ÅÁöÑËæìÂÖ•„ÄÇ
# model(**tokens) Ë∞ÉÁî®Ê®°ÂûãÂØπËæìÂÖ•ÁöÑÂàÜËØçÁªìÊûúËøõË°åÂ§ÑÁêÜÔºåÊ®°Âûã‰ºöÊ†πÊçÆËæìÂÖ•ÁöÑËØçÂÖÉ ID ËøõË°åÂâçÂêë‰º†Êí≠ËÆ°ÁÆó„ÄÇ
```

# Recommending songs by embeddings

```python
import pandas as pd
from urllib import request

# Get the playlist dataset file ËØ•ÂáΩÊï∞‰ºöÂêëÊåáÂÆöÁöÑ URL ÂèëÈÄÅ HTTP ËØ∑Ê±ÇÔºåÂπ∂ËøîÂõû‰∏Ä‰∏™ÂìçÂ∫îÂØπË±°„ÄÇËøôÈáåÁöÑ URL ÊåáÂêë‰∏Ä‰∏™Êí≠ÊîæÂàóË°®Êï∞ÊçÆÈõÜÊñá‰ª∂ train.txt„ÄÇ
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')

# Parse the playlist dataset file. Skip the first two lines as
# they only contain metadata
lines = data.read().decode("utf-8").split('\n')[2:]
# data.read()ÔºöËØªÂèñÂìçÂ∫îÂØπË±° data ‰∏≠ÁöÑÂÜÖÂÆπÔºåËøîÂõû‰∏Ä‰∏™Â≠óËäÇÁ±ªÂûãÁöÑÊï∞ÊçÆ„ÄÇ
# .decode("utf-8")ÔºöÂ∞ÜÂ≠óËäÇÊï∞ÊçÆËß£Á†Å‰∏∫ UTF - 8 ÁºñÁ†ÅÁöÑÂ≠óÁ¨¶‰∏≤„ÄÇ
# .split('\n')ÔºöÂ∞ÜËß£Á†ÅÂêéÁöÑÂ≠óÁ¨¶‰∏≤ÊåâÊç¢Ë°åÁ¨¶ \n ËøõË°åÂàÜÂâ≤ÔºåÂæóÂà∞‰∏Ä‰∏™Â≠óÁ¨¶‰∏≤ÂàóË°®ÔºåÊØè‰∏™ÂÖÉÁ¥†‰ª£Ë°®Êñá‰ª∂‰∏≠ÁöÑ‰∏ÄË°å„ÄÇ
# [2:]ÔºöÂàáÁâáÊìç‰ΩúÔºåË∑≥ËøáÂàóË°®‰∏≠ÁöÑÂâç‰∏§‰∏™ÂÖÉÁ¥†ÔºåÂõ†‰∏∫Ââç‰∏§Ë°åÂè™ÂåÖÂê´ÂÖÉÊï∞ÊçÆÔºåÊàë‰ª¨Âè™ÈúÄË¶ÅÂÆûÈôÖÁöÑÊí≠ÊîæÂàóË°®Êï∞ÊçÆ„ÄÇ

# Remove playlists with only one song
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]
# s.rstrip()ÔºöÂØπÂàóË°® lines ‰∏≠ÁöÑÊØè‰∏™Â≠óÁ¨¶‰∏≤ s ËøõË°åÂ§ÑÁêÜÔºåÂéªÈô§Â≠óÁ¨¶‰∏≤Êú´Â∞æÁöÑÁ©∫ÁôΩÂ≠óÁ¨¶„ÄÇ
# .split()ÔºöÂ∞ÜÂ§ÑÁêÜÂêéÁöÑÂ≠óÁ¨¶‰∏≤ÊåâÁ©∫ÁôΩÂ≠óÁ¨¶ËøõË°åÂàÜÂâ≤ÔºåÂæóÂà∞‰∏Ä‰∏™Áî±Ê≠åÊõ≤ ID ÁªÑÊàêÁöÑÂàóË°®Ôºå‰ª£Ë°®‰∏Ä‰∏™Êí≠ÊîæÂàóË°®„ÄÇ
# if len(s.split()) > 1ÔºöËøáÊª§Êù°‰ª∂ÔºåÂè™‰øùÁïôÂåÖÂê´Â§ö‰∫é‰∏ÄÈ¶ñÊ≠åÊõ≤ÁöÑÊí≠ÊîæÂàóË°®

# Load song metadata
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file]
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist'])
songs_df = songs_df.set_index('id')
```

```python
from gensim.models import Word2Vec
# gensim ÊòØ‰∏Ä‰∏™Áî®‰∫é‰∏ªÈ¢òÂª∫Ê®°„ÄÅÊñáÊ°£Á¥¢ÂºïÂíåÁõ∏‰ººÊÄßÊ£ÄÁ¥¢ÁöÑ Python Â∫ìÔºåÂú®Ëá™ÁÑ∂ËØ≠Ë®ÄÂ§ÑÁêÜÈ¢ÜÂüüÂ∫îÁî®ÂπøÊ≥õ„ÄÇ
# Word2Vec ÊòØ gensim Â∫ì‰∏≠ÁöÑ‰∏Ä‰∏™Á±ªÔºåÁî®‰∫éËÆ≠ÁªÉËØçÂêëÈáèÊ®°Âûã„ÄÇËØçÂêëÈáèÊ®°ÂûãÂèØ‰ª•Â∞ÜÂçïËØçË°®Á§∫‰∏∫ÂêëÈáèÔºå‰ΩøÂæóËØ≠‰πâÁõ∏ËøëÁöÑÂçïËØçÂú®ÂêëÈáèÁ©∫Èó¥‰∏≠Ë∑ùÁ¶ªËæÉËøëÔºå‰ªéËÄå‰æø‰∫éËøõË°åËØ≠‰πâÂàÜÊûê„ÄÅÊñáÊú¨ÂàÜÁ±ªÁ≠â‰ªªÂä°„ÄÇ

# Train our Word2Vec model
model = Word2Vec(
    playlists, vector_size=32, window=20, negative=50, min_count=1, workers=4
)
# playlistsÔºöËøôÊòØËÆ≠ÁªÉÊï∞ÊçÆÔºåÈÄöÂ∏∏ÊòØ‰∏Ä‰∏™ÂàóË°®ÁöÑÂàóË°®„ÄÇËøôÈáå playlists ÊòØ‰πãÂâç‰ª£Á†Å‰∏≠Â§ÑÁêÜÂæóÂà∞ÁöÑÊí≠ÊîæÂàóË°®Êï∞ÊçÆÔºåÊØè‰∏™Â≠êÂàóË°®‰ª£Ë°®‰∏Ä‰∏™Êí≠ÊîæÂàóË°®ÔºåÂàóË°®‰∏≠ÁöÑÂÖÉÁ¥†ÊòØÊ≠åÊõ≤ÁöÑ ID„ÄÇWord2Vec Ê®°Âûã‰ºöÂ∞ÜËøô‰∫õÊ≠åÊõ≤ ID ÂΩì‰Ωú ‚ÄúÂçïËØç‚Äù Êù•Â§ÑÁêÜÔºåÂ≠¶‰π†ÂÆÉ‰ª¨‰πãÈó¥ÁöÑËØ≠‰πâÂÖ≥Á≥ª„ÄÇ
# vector_size=32ÔºöÊåáÂÆöËØçÂêëÈáèÁöÑÁª¥Â∫¶„ÄÇËøôÈáåÂ∞ÜÊØè‰∏™Ê≠åÊõ≤ ID ÂØπÂ∫îÁöÑÂêëÈáèÁª¥Â∫¶ËÆæÁΩÆ‰∏∫ 32„ÄÇÁª¥Â∫¶Ë∂äÈ´òÔºåÊ®°ÂûãËÉΩÂ§üÊçïÊçâÁöÑËØ≠‰πâ‰ø°ÊÅØË∂ä‰∏∞ÂØåÔºå‰ΩÜËÆ≠ÁªÉÊó∂Èó¥ÂíåÂÜÖÂ≠òÊ∂àËÄó‰πü‰ºöÁõ∏Â∫îÂ¢ûÂä†„ÄÇ
# window=20ÔºöË°®Á§∫ËÆ≠ÁªÉÊó∂ÁöÑ‰∏ä‰∏ãÊñáÁ™óÂè£Â§ßÂ∞è„ÄÇÂú®Â§ÑÁêÜÊØè‰∏™ ‚ÄúÂçïËØç‚ÄùÔºàÊ≠åÊõ≤ IDÔºâÊó∂Ôºå‰ºöËÄÉËôëÂÖ∂ÂâçÂêé window ‰∏™ ‚ÄúÂçïËØç‚Äù ‰Ωú‰∏∫‰∏ä‰∏ãÊñá„ÄÇ‰æãÂ¶ÇÔºåÂΩì window = 20 Êó∂ÔºåÂØπ‰∫éÊüê‰∏™Ê≠åÊõ≤ IDÔºå‰ºöËÄÉËôëÂÆÉÂâçÂêéÂêÑ 20 ‰∏™Ê≠åÊõ≤ ID ‰Ωú‰∏∫‰∏ä‰∏ãÊñá‰ø°ÊÅØÊù•Â≠¶‰π†ÂÖ∂ÂêëÈáèË°®Á§∫„ÄÇ
# negative=50ÔºöË¥üÈááÊ†∑ÁöÑÊï∞Èáè„ÄÇË¥üÈááÊ†∑ÊòØ‰∏ÄÁßç‰ºòÂåñËÆ≠ÁªÉÊïàÁéáÁöÑÊñπÊ≥ïÔºåÂú®ËÆ≠ÁªÉËøáÁ®ã‰∏≠Ôºå‰ºöÈöèÊú∫ÈÄâÊã© negative ‰∏™ ‚ÄúË¥üÊ†∑Êú¨‚Äù Êù•Êõ¥Êñ∞Ê®°ÂûãÂèÇÊï∞ÔºåÈÅøÂÖçÂØπÊï¥‰∏™ËØçÊ±áË°®ËøõË°åÊõ¥Êñ∞Ôºå‰ªéËÄåÂä†Âø´ËÆ≠ÁªÉÈÄüÂ∫¶„ÄÇ
# min_count=1ÔºöËÆæÁΩÆÊúÄÂ∞èËØçÈ¢ë„ÄÇÂè™ÊúâÂá∫Áé∞Ê¨°Êï∞Â§ß‰∫éÁ≠â‰∫é min_count ÁöÑ ‚ÄúÂçïËØç‚ÄùÔºàÊ≠åÊõ≤ IDÔºâÊâç‰ºöË¢´Á∫≥ÂÖ•Ê®°ÂûãÁöÑËÆ≠ÁªÉËåÉÂõ¥„ÄÇËøôÈáåËÆæÁΩÆ‰∏∫ 1ÔºåË°®Á§∫ÊâÄÊúâÂá∫Áé∞ËøáÁöÑÊ≠åÊõ≤ ID ÈÉΩ‰ºöË¢´ËÄÉËôë„ÄÇ
# workers=4ÔºöÊåáÂÆöËÆ≠ÁªÉÊó∂‰ΩøÁî®ÁöÑÁ∫øÁ®ãÊï∞„ÄÇÂ§öÁ∫øÁ®ãÂèØ‰ª•Âπ∂Ë°åÂ§ÑÁêÜËÆ≠ÁªÉÊï∞ÊçÆÔºåÂä†Âø´ËÆ≠ÁªÉÈÄüÂ∫¶„ÄÇËøôÈáåËÆæÁΩÆ‰∏∫ 4ÔºåË°®Á§∫‰ΩøÁî® 4 ‰∏™Á∫øÁ®ãËøõË°åËÆ≠ÁªÉ„ÄÇ
```

```python
import numpy as np

def print_recommendations(song_id):
    similar_songs = np.array(
        model.wv.most_similar(positive=str(song_id),topn=5)
    )[:,0]
    return  songs_df.iloc[similar_songs]
# model.wv ÊòØ Word2Vec Ê®°ÂûãÁöÑËØçÂêëÈáèÂØπË±°ÔºåÁî®‰∫éËÆøÈóÆËÆ≠ÁªÉÂ•ΩÁöÑËØçÂêëÈáè„ÄÇ
# most_similar ÊñπÊ≥ïÁî®‰∫éÊâæÂá∫‰∏éÁªôÂÆö ‚ÄúÂçïËØç‚ÄùÔºàËøôÈáåÊòØÊ≠åÊõ≤ IDÔºâÊúÄÁõ∏‰ººÁöÑÂÖ∂‰ªñ ‚ÄúÂçïËØç‚Äù„ÄÇ
# positive ÂèÇÊï∞ÊåáÂÆöË¶ÅÊü•ËØ¢ÁöÑÁõÆÊ†á ‚ÄúÂçïËØç‚ÄùÔºåËøôÈáåÂ∞Ü song_id ËΩ¨Êç¢‰∏∫Â≠óÁ¨¶‰∏≤Á±ªÂûãÂêé‰º†ÂÖ•„ÄÇ
# topn=5 Ë°®Á§∫ËøîÂõûÊúÄÁõ∏‰ººÁöÑÂâç 5 ‰∏™ ‚ÄúÂçïËØç‚ÄùÔºàÊ≠åÊõ≤ IDÔºâÂèäÂÖ∂Áõ∏‰ººÂ∫¶ÂæóÂàÜ„ÄÇËØ•ÊñπÊ≥ïËøîÂõû‰∏Ä‰∏™ÂåÖÂê´Â§ö‰∏™ÂÖÉÁªÑÁöÑÂàóË°®ÔºåÊØè‰∏™ÂÖÉÁªÑÁöÑÁ¨¨‰∏Ä‰∏™ÂÖÉÁ¥†ÊòØÁõ∏‰ººÊ≠åÊõ≤ÁöÑ IDÔºåÁ¨¨‰∫å‰∏™ÂÖÉÁ¥†ÊòØÁõ∏‰ººÂ∫¶ÂæóÂàÜ„ÄÇ
# Extract recommendations
print_recommendations(2172)
```
