# Downloading and Running An LLM

The first step is to load our model onto the GPU for faster inference. Note that we load the model and tokenizer separately and keep them as such so that we can explore them separately.

```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# Load model and tokenizer
model = AutoModelForCausalLM.from_pretrained(
    "microsoft/Phi-3-mini-4k-instruct",
    device_map="cuda",
    torch_dtype="auto",
    trust_remote_code=True,
)
tokenizer = AutoTokenizer.from_pretrained("microsoft/Phi-3-mini-4k-instruct")
```

```python
prompt = "Write an email apologizing to Sarah for the tragic gardening mishap. Explain how it happened.<|assistant|>"

# Tokenize the input prompt
input_ids = tokenizer(prompt, return_tensors="pt").input_ids.to("cuda")
# (prompt, return_tensors="pt")：调用分词器对 prompt 进行分词处理，return_tensors="pt" 表示返回 PyTorch 张量。
# .input_ids：从分词器的输出中提取 input_ids，即经过分词后得到的数字序列。

# Generate the text
generation_output = model.generate(
  input_ids=input_ids, # 指定模型的输入，即之前经过分词处理得到的数字序列。
  max_new_tokens=20 # 设置模型最多生成 x tokens，但是似乎不能保证这段话能结尾
)

# Print the output
print(tokenizer.decode(generation_output[0])) # tokenizer.decode：将生成的数字序列（input_ids）转换回文本字符串。
```

# Comparing Trained LLM Tokenizers


```python
from transformers import AutoModelForCausalLM, AutoTokenizer

# colors_list 是一个字符串列表，每个字符串代表一种 RGB 颜色值。这些颜色值将用于在终端中为不同的词元添加背景颜色，以达到可视化的效果。
colors_list = [
    '102;194;165', '252;141;98', '141;160;203',
    '231;138;195', '166;216;84', '255;217;47'
]

def show_tokens(sentence, tokenizer_name):
    tokenizer = AutoTokenizer.from_pretrained(tokenizer_name)
    token_ids = tokenizer(sentence).input_ids
    for idx, t in enumerate(token_ids):
        print(
            f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m' +
            tokenizer.decode(t) +
            '\x1b[0m',
            end=' '
        )
  #  f'\x1b[0;30;48;2;{colors_list[idx % len(colors_list)]}m'是一个 ANSI 转义序列，用于设置终端文本的样式。
  #  其中 0 表示重置所有属性，30 表示前景色为黑色，48;2 表示使用 RGB 颜色设置背景色，
  #  {colors_list[idx % len(colors_list)]} 是从 colors_list 中循环选取的 RGB 颜色值。
  #  '\x1b[0m'：另一个 ANSI 转义序列，用于重置终端文本的样式，确保后续输出不受之前样式的影响。
  #  end=' '：设置 print 函数的结束符为空格，使得每个词元在同一行输出，用空格分隔。
```

```python
text = """
English and CAPITALIZATION
🎵 鸟
show_tokens False None elif == >= else: two tabs:"    " Three tabs: "       "
12.0*50=600
""" # 这里有各种tokens
```

```python
show_tokens(text, "bert-base-uncased") # 不区分大小写
```

```python
show_tokens(text, "bert-base-cased") # 区分大小写
```

# Contextualized Word Embeddings From a Language Model (Like BERT)
# 来自语言模型的上下文化词嵌入（如 BERT）

```python
from transformers import AutoModel, AutoTokenizer

# Load a tokenizer
tokenizer = AutoTokenizer.from_pretrained("microsoft/deberta-base")

# Load a language model
model = AutoModel.from_pretrained("microsoft/deberta-v3-xsmall")

# Tokenize the sentence
tokens = tokenizer('Hello world', return_tensors='pt') # return_tensors='pt' 表示返回 PyTorch 张量格式的分词结果。

# Process the tokens
output = model(**tokens)[0]
# **tokens 是 Python 中的解包操作，将 tokens 字典中的键值对解包为关键字参数传递给 model。这样，model 就可以接收 input_ids、attention_mask 等必要的输入。
# model(**tokens) 调用模型对输入的分词结果进行处理，模型会根据输入的词元 ID 进行前向传播计算。
```

# Recommending songs by embeddings

```python
import pandas as pd
from urllib import request

# Get the playlist dataset file 该函数会向指定的 URL 发送 HTTP 请求，并返回一个响应对象。这里的 URL 指向一个播放列表数据集文件 train.txt。
data = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/train.txt')

# Parse the playlist dataset file. Skip the first two lines as
# they only contain metadata
lines = data.read().decode("utf-8").split('\n')[2:]
# data.read()：读取响应对象 data 中的内容，返回一个字节类型的数据。
# .decode("utf-8")：将字节数据解码为 UTF - 8 编码的字符串。
# .split('\n')：将解码后的字符串按换行符 \n 进行分割，得到一个字符串列表，每个元素代表文件中的一行。
# [2:]：切片操作，跳过列表中的前两个元素，因为前两行只包含元数据，我们只需要实际的播放列表数据。

# Remove playlists with only one song
playlists = [s.rstrip().split() for s in lines if len(s.split()) > 1]
# s.rstrip()：对列表 lines 中的每个字符串 s 进行处理，去除字符串末尾的空白字符。
# .split()：将处理后的字符串按空白字符进行分割，得到一个由歌曲 ID 组成的列表，代表一个播放列表。
# if len(s.split()) > 1：过滤条件，只保留包含多于一首歌曲的播放列表

# Load song metadata
songs_file = request.urlopen('https://storage.googleapis.com/maps-premium/dataset/yes_complete/song_hash.txt')
songs_file = songs_file.read().decode("utf-8").split('\n')
songs = [s.rstrip().split('\t') for s in songs_file]
songs_df = pd.DataFrame(data=songs, columns = ['id', 'title', 'artist'])
songs_df = songs_df.set_index('id')
```

```python
print( 'Playlist #1:\n ', playlists[0], '\n')
print( 'Playlist #2:\n ', playlists[1])
```

```python
from gensim.models import Word2Vec
# gensim 是一个用于主题建模、文档索引和相似性检索的 Python 库，在自然语言处理领域应用广泛。
# Word2Vec 是 gensim 库中的一个类，用于训练词向量模型。词向量模型可以将单词表示为向量，使得语义相近的单词在向量空间中距离较近，从而便于进行语义分析、文本分类等任务。

# Train our Word2Vec model
model = Word2Vec(
    playlists, vector_size=32, window=20, negative=50, min_count=1, workers=4
)
# playlists：这是训练数据，通常是一个列表的列表。这里 playlists 是之前代码中处理得到的播放列表数据，每个子列表代表一个播放列表，列表中的元素是歌曲的 ID。Word2Vec 模型会将这些歌曲 ID 当作 “单词” 来处理，学习它们之间的语义关系。
# vector_size=32：指定词向量的维度。这里将每个歌曲 ID 对应的向量维度设置为 32。维度越高，模型能够捕捉的语义信息越丰富，但训练时间和内存消耗也会相应增加。
# window=20：表示训练时的上下文窗口大小。在处理每个 “单词”（歌曲 ID）时，会考虑其前后 window 个 “单词” 作为上下文。例如，当 window = 20 时，对于某个歌曲 ID，会考虑它前后各 20 个歌曲 ID 作为上下文信息来学习其向量表示。
# negative=50：负采样的数量。负采样是一种优化训练效率的方法，在训练过程中，会随机选择 negative 个 “负样本” 来更新模型参数，避免对整个词汇表进行更新，从而加快训练速度。
# min_count=1：设置最小词频。只有出现次数大于等于 min_count 的 “单词”（歌曲 ID）才会被纳入模型的训练范围。这里设置为 1，表示所有出现过的歌曲 ID 都会被考虑。
# workers=4：指定训练时使用的线程数。多线程可以并行处理训练数据，加快训练速度。这里设置为 4，表示使用 4 个线程进行训练。
```

```python
song_id = 2172

# Ask the model for songs similar to song #2172
model.wv.most_similar(positive=str(song_id))
```

```python
import numpy as np

def print_recommendations(song_id):
    similar_songs = np.array(
        model.wv.most_similar(positive=str(song_id),topn=5)
    )[:,0]
    return  songs_df.iloc[similar_songs]
# model.wv 是 Word2Vec 模型的词向量对象，用于访问训练好的词向量。
# most_similar 方法用于找出与给定 “单词”（这里是歌曲 ID）最相似的其他 “单词”。
# positive 参数指定要查询的目标 “单词”，这里将 song_id 转换为字符串类型后传入。
# topn=5 表示返回最相似的前 5 个 “单词”（歌曲 ID）及其相似度得分。该方法返回一个包含多个元组的列表，每个元组的第一个元素是相似歌曲的 ID，第二个元素是相似度得分。
# Extract recommendations
print_recommendations(2172)
```
